```python
class Node:
    """
    链表结点类。

    每个结点包含：
    - data：存储的数据
    - next_node：指向下一个结点的引用（链）
    """
    def __init__(self, data: str | int | float):
        """
        初始化结点。

        :param data: 结点存储的数据
        """
        self.data: str | int | float = data
        self.next_node: Node | None = None

    def __repr__(self) -> str:
        """返回结点的字符串表示。"""
        return f"Node({self.data!r})"


class LinkedList:
    """
    链表类。

    维护一个指向第一个结点的头指针，
    通过头指针可以遍历整个链表。
    """
    def __init__(self):
        """初始化空链表。"""
        self.first_node: Node | None = None

    def __repr__(self) -> str:
        """返回链表的字符串表示。"""
        if not self.first_node:
            return "LinkedList([])"

        # 初始化空列表
        nodes: list[str] = []
        # 指定链表的查询起始点
        # 可以不在类初始化时完成
        current: Node | None = self.first_node

        # 从第一个Node开始
        while current:
            nodes.append(str(current.data))
            # 如果有下一个节点，while循环继续
            current = current.next_node

        return f"LinkedList([{', '.join(nodes)}])"
```

```python
# 创建链表: "once" -> "upon" -> "a" -> "time"

# 创建结点
node_1 = Node("once")
node_2 = Node("upon")
node_3 = Node("a")
node_4 = Node("time")

# 连接结点
# 注意: Node.next_node应该是一个Node，而不是str
node_1.next_node = node_2
node_2.next_node = node_3
node_3.next_node = node_4

# 创建链表，指定头结点
list_ = LinkedList()
list_.first_node = node_1

# 打印链表对象（触发__repr__）
print(f"链表: {list_}")

print("\n遍历链表:")
# 从头结点开始遍历
current = list_.first_node
while current:
    print(f" {current.data}")
    current = current.next_node
```

操作1：读取

数组：O(1)
链表：O(N)

```python
def read(self, index: int) -> str | int | float | None:
    """
    读取指定索引位置的元素。

    时间复杂度：O(N)

    :param index: 要读取的索引（从0开始）
    :return: 索引位置的元素值，索引越界返回None
    """
    current_node: Node | None = self.first_node
    current_index: int = 0

    # 有了索引，仍然需要遍历才能到达索引位置
    while current_index < index:
        if not current_node:
            return None  # 索引越界，无法在遍历之前就发现越界，没有分配整段内存
        current_node = current_node.next_node
        current_index += 1

    # current_index += 1在遍历最后一次结束前执行过
    return current_node.data if current_node else None

# 绑定方法到类
LinkedList.read = read

# 测试读取
print(f"链表: {list_}")
print(f"\n读取索引0: {list_.read(0)}")  # "once"
print(f"读取索引1: {list_.read(1)}")  # "upon"
print(f"读取索引2: {list_.read(2)}")  # "a"
print(f"读取索引3: {list_.read(3)}")  # "time"
print(f"读取索引10: {list_.read(10)}")  # None（越界）
```

操作2：查找

数组：O(N)
链表：O(N)

```python
def index_of(self, value: str | int | float) -> int | None:
    """
    查找指定值的索引。

    时间复杂度：O(N)

    :param value: 要查找的值
    :return: 值的索引，未找到返回None
    """
    current_node: Node | None = self.first_node
    current_index: int = 0

    # 如果没有对应元素，则从头查到尾
    while current_node:
        if current_node.data == value:
            return current_index
        current_node = current_node.next_node
        current_index += 1

    return None  # 未找到


# 绑定方法
LinkedList.index_of = index_of
```

操作3：插入

需要先找到插入位置 → 遍历 N 个结点 → O (N)

```python
def insert_at_index(self, index: int, value: str | int | float) -> bool:
    """
    在指定索引位置插入新元素。

    时间复杂度:
    - 索引0（开头）：O(1)
    - 其他位置：O(N)

    :param index: 插入位置索引
    :param value: 要插入的值
    :return: 插入成功返回True，索引越界返回False
    """
    new_node = Node(value)

    # 在开头插入
    if index == 0:
        new_node.next_node = self.first_node
        self.first_node = new_node
        return True

    # 准备好查询节点
    current_node: Node | None = self.first_node
    current_index: int = 0

    # 找到插入位置的前一个结点
    # ["hello"] -->  ["world"] --> ["! "]
    #   0             1            2
    # 每一轮while循环向右移动一个node
    # while循环结束的时候正好停在index - 1
    while current_index < index - 1:
        if not current_node:
            return False  # 索引越界
        current_node = current_node.next_node
        current_index += 1

    # 循环结束后补充检查一次索引越界
    if not current_node:
        return False
    # 插入新结点
    # 就像插队一样
    new_node.next_node = current_node.next_node
    current_node.next_node = new_node
    return True


# 绑定方法
LinkedList.insert_at_index = insert_at_index

# 测试插入
test_list = LinkedList()
test_list.insert_at_index(0, "world")    # 开头插入
test_list.insert_at_index(0, "hello")   # 开头插入
test_list.insert_at_index(2, "!")       # 末尾插入
test_list.insert_at_index(2, " beautiful") # 中间插入

print(f"插入后: {test_list}")
```

操作4：删除

```python
def remove_at_index(self, index: int) -> bool:
    """
    删除指定索引位置的节点。

    时间复杂度:
    - 索引0（开头）：O(1)
    - 其他位置：O(N)

    :param index: 删除位置索引
    :return: 删除成功返回True，索引越界返回False
    """
    # 链表为空，直接返回失败
    if not self.first_node:
        return False

    # 删除开头节点
    if index == 0:
        self.first_node = self.first_node.next_node
        return True

    # 准备好查询节点
    current_node: Node | None = self.first_node
    current_index: int = 0

    # 找到删除位置的前一个结点
    # 每一轮while循环向右移动一个node
    # while循环结束的时候正好停在index - 1
    while current_index < index - 1:
        if not current_node:
            return False  # 索引越界
        current_node = current_node.next_node
        current_index += 1

    # 循环结束后补充检查索引越界
    if not current_node or not current_node.next_node:
        return False

    # 删除目标节点
    # 跳过要删除的节点，直接指向下下个节点
    current_node.next_node = current_node.next_node.next_node
    return True


# 绑定方法
LinkedList.remove_at_index = remove_at_index

# 测试删除
test_list = LinkedList()
test_list.insert_at_index(0, "world")
test_list.insert_at_index(0, "hello")
test_list.insert_at_index(2, "!")
test_list.insert_at_index(2, " beautiful")

print(f"删除前: {test_list}")
test_list.remove_at_index(2)  # 删除中间
test_list.remove_at_index(0)  # 删除开头
test_list.remove_at_index(1)  # 删除末尾
print(f"删除后: {test_list}")
```
