## 作业

1. **实现链表反转**：编写一个方法 `reverse()`，将链表反转。例如：`[a, b, c]` → `[c, b, a]`

2. **检测环**：编写一个方法 `has_cycle()`，判断链表中是否存在环（某个结点的next指向了前面的结点）。

```python
class DNode:
    """双向链表结点"""
    def __init__(self, data: str | int | float):
        self.data = data
        self.prev: DNode | None = None
        self.next: DNode | None = None

class DoublyLinkedList:
    """双向链表"""
    def __init__(self):
        self.first: DNode | None = None
        self.last: DNode | None = None
    
    def insert_at_end(self, value: str | int | float) -> None:
        """在末尾插入 - O(1)"""
        new_node = DNode(value)
        if not self.first:  # 空链表
            self.first = self.last = new_node
        else:
            # 新节点的prev指向原来的队尾
            new_node.prev = self.last
            # 原队尾的next指向新节点
            self.last.next = new_node
            # 新队尾指向新节点
            self.last = new_node
    
    def remove_from_front(self) -> str | int | float | None:
        """从开头删除 - O(1)"""
        if not self.first:
            return None
        data = self.first.data
        self.first = self.first.next
        if self.first:
            self.first.prev = None
        else:
            self.last = None
        return data
    
    def reverse(self) -> 'DoublyLinkedList':
        """将列表反转 - O(n)"""
        # 空链表或只有一个节点，直接返回
        if not self.first or self.first == self.last:
            return self
        
        current = self.first
        # 遍历所有节点，交换 prev 和 next
        while current is not None:
            # 保存原来的 prev，因为交换后会被覆盖
            temp = current.prev
            # 交换 prev 和 next
            current.prev = current.next
            current.next = temp
            # 移动到下一个节点（原来的 next 现在是 prev）
            current = current.prev
        
        # 最后交换 first 和 last 指针
        self.first, self.last = self.last, self.first
        return self
    
    def has_cycle(self) -> bool:
        """判断环形双向链表是否有环"""
        # 空链表一定无环
        if not self.first or self.first == self.last:
            return False

        current = self.first
        my_list=[]  
        while current not in my_list:
            my_list.append(current)
            if current==self.first:
                return False
            current = current.next
        return True


dll = DoublyLinkedList()
dll.insert_at_end(1)
dll.insert_at_end(2)
dll.insert_at_end(3)

# 反转
dll.reverse()

# 验证：3 → 2 → 1
print(dll.remove_from_front())  # 3
print(dll.remove_from_front())  # 2
print(dll.remove_from_front())  # 1


# 验证是否有环
print(dll.has_cycle())  # 输出：True
```
